```python
import yfinance as yf
import pandas as pd
```

```python
# Get market value of all companies

companies = [
    yf.Ticker(ticker).info for ticker in tickers
]

df = pd.DataFrame(companies)
df = df[["symbol", "marketCap"]]
df = df.dropna()
df = df.sort_values(by="marketCap", ascending=False)
df = df.reset_index(drop=True)
df
```

```python
all_tickers = pd.read_csv("../data/tickers.csv")["Symbol"]

all_companies = [
    yf.Ticker(ticker).info for ticker in all_tickers
]

all_df = pd.DataFrame(all_companies)
all_df = all_df[["symbol", "marketCap"]]
all_df = all_df.dropna()
all_df = all_df.sort_values(by="marketCap", ascending=False)
all_df = all_df.reset_index(drop=True)
all_df
```

```python
# Calculate the total of the market value of the companies in the S&P 500

sparse_market_cap = df["marketCap"].sum()

market_cap = all_df["marketCap"].sum()

sparse_market_cap / market_cap
```
